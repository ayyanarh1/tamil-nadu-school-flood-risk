```python
# Cell 2 — Fixed, no header line
from fpdf import FPDF

class Report(FPDF):
    def header(self):
        pass  # No header — removes the line and text

    def footer(self):
        self.set_y(-15)
        self.set_font('Helvetica', 'I', 8)
        self.set_text_color(128)
        self.cell(0, 10, 'Page ' + str(self.page_no()), align='C')

    def section_title(self, title):
        self.set_font('Helvetica', 'B', 13)
        self.set_text_color(6, 90, 130)
        self.set_fill_color(232, 244, 253)
        self.cell(0, 9, title, fill=True)
        self.ln(5)
        self.set_text_color(0)

    def body_text(self, text):
        self.set_font('Helvetica', '', 11)
        self.set_text_color(50)
        self.multi_cell(0, 6, text)
        self.ln(3)

    def bullet(self, text):
        self.set_font('Helvetica', '', 11)
        self.set_text_color(50)
        self.set_x(15)
        self.multi_cell(0, 6, '- ' + text)

pdf = Report()
pdf.set_auto_page_break(auto=True, margin=15)
pdf.add_page()

# Title
pdf.set_font('Helvetica', 'B', 20)
pdf.set_text_color(2, 28, 59)
pdf.cell(0, 12, 'Tamil Nadu School Flood Risk', align='C')
pdf.ln(8)
pdf.set_font('Helvetica', 'B', 14)
pdf.set_text_color(6, 90, 130)
pdf.cell(0, 8, 'Satellite-Based Climate Hazard Assessment', align='C')
pdf.ln(6)
pdf.set_font('Helvetica', '', 11)
pdf.set_text_color(100)
pdf.cell(0, 6, 'UNICEF Digital Inclusion - Giga Initiative | March 2026', align='C')
pdf.ln(10)

# Section 1
pdf.section_title('1. Executive Summary')
pdf.body_text('Satellite-based flood risk assessment for 15 schools across Tamil Nadu combining five data sources spanning 40 years of climate and flood observations.')
pdf.body_text('Two schools are at CRITICAL risk. Nagapattinam has no connectivity - a flood event cuts both physical access and digital learning simultaneously.')
pdf.body_text('Under SSP5-8.5 worst case, five schools escalate to CRITICAL by 2050.')

# Section 2
pdf.section_title('2. Data Sources')
pdf.bullet('Sentinel-1 SAR 2023 - flood pixels via backscatter threshold')
pdf.bullet('JRC Surface Water 1984-2021 - 37 years of water occurrence')
pdf.bullet('Global Flood Database 2000-2018 - 913 major flood events')
pdf.bullet('ERA5 Rainfall 2018-2023 - monthly precipitation per school')
pdf.bullet('CMIP6 SSP projections - 2050 delta method scenarios')
pdf.ln(4)

# Section 3
pdf.section_title('3. Risk Scoring')
pdf.body_text('Formula: JRC(50%) + SAR(30%) + GFD(20%) + ERA5(10%) + Connectivity penalty(+15)')
pdf.bullet('CRITICAL: score >= 50')
pdf.bullet('HIGH: score 25-50')
pdf.bullet('MEDIUM: score 10-25')
pdf.bullet('LOW: score < 10')
pdf.ln(4)

# Section 4
pdf.add_page()
pdf.section_title('4. Key Findings')
pdf.body_text('CRITICAL: School Puducherry Border (96.7) - Connected')
pdf.body_text('CRITICAL: Panchayat School Nagapattinam (56.8) - NO CONNECTIVITY')
pdf.ln(3)
pdf.body_text('HIGH risk schools:')
pdf.bullet('Ramanathapuram (43.5) - No connectivity')
pdf.bullet('Cuddalore (39.8) - No connectivity')
pdf.bullet('Tuticorin (39.1) - Connected')
pdf.bullet('Kanchipuram (33.3) - No connectivity')
pdf.bullet('Tiruchirappalli (31.4) - No connectivity')
pdf.bullet('Tirunelveli (30.7) - No connectivity')
pdf.bullet('Thanjavur (26.5) - Connected')
pdf.ln(4)
pdf.body_text('LOW risk (safe in all scenarios):')
pdf.bullet('High School Coimbatore - zero flood history')
pdf.bullet('Govt School Salem - zero flood history')
pdf.ln(4)

# Section 5
pdf.section_title('5. 2050 Climate Projections')
pdf.body_text('SSP2-4.5 (+15% rainfall): 3 schools escalate - Ramanathapuram reaches CRITICAL')
pdf.body_text('SSP5-8.5 (+30% rainfall): 5 schools escalate - Ramanathapuram, Cuddalore, Tuticorin reach CRITICAL')
pdf.ln(4)

# Section 6
pdf.section_title('6. Recommendations')
pdf.bullet('Immediate: Connectivity investment for Nagapattinam and Ramanathapuram')
pdf.bullet('Monitor: Sentinel-1 real-time alerts Oct-Dec for coastal schools')
pdf.bullet('Plan: Integrate 2050 SSP projections into infrastructure planning')
pdf.bullet('Expand: Apply to full Giga dataset - 5000+ schools')
pdf.ln(6)

pdf.set_font('Helvetica', 'I', 10)
pdf.set_text_color(100)
pdf.cell(0, 6, 'github.com/ayyanarh1/tamil-nadu-school-flood-risk', align='C')

# Save
pdf.output('tamil_nadu_flood_risk_report.pdf')
print('PDF saved successfully!')
print('Pages:', pdf.page_no())
```
