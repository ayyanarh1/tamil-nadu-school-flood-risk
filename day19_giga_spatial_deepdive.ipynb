```python
# Cell 7 — Create our AQUEDUCT flood handler
flood_handler_code = '''"""
AQUEDUCT Flood Hazard Handler for Giga Spatial

Integrates WRI AQUEDUCT flood hazard layers with school location data
to assess flood risk at school level.

Data source: World Resources Institute AQUEDUCT 3.0
URL: https://www.wri.org/aqueduct
"""

import numpy as np
import pandas as pd
import geopandas as gpd
import requests
import rasterio
from rasterio.mask import mask
from pathlib import Path
from typing import Optional, Union, List, Literal
from pydantic import ConfigDict
from pydantic.dataclasses import dataclass
from shapely.geometry import Point, mapping
import tempfile
import os
import logging

logger = logging.getLogger(__name__)


# AQUEDUCT return period options
RETURN_PERIODS = {
    "rp10":   "1-in-10 year flood (10% annual probability)",
    "rp25":   "1-in-25 year flood (4% annual probability)",
    "rp50":   "1-in-50 year flood (2% annual probability)",
    "rp100":  "1-in-100 year flood (1% annual probability)",
    "rp250":  "1-in-250 year flood (0.4% annual probability)",
    "rp500":  "1-in-500 year flood (0.2% annual probability)",
    "rp1000": "1-in-1000 year flood (0.1% annual probability)",
}

# AQUEDUCT hazard types
HAZARD_TYPES = {
    "inunriver": "Riverine flooding",
    "inuncoast": "Coastal flooding",
}


@dataclass(config=ConfigDict(arbitrary_types_allowed=True))
class AQUEDUCTFloodHandler:
    """
    Handler for WRI AQUEDUCT 3.0 flood hazard data.

    Extracts flood depth values at school locations for multiple
    return periods, enabling flood risk assessment for education
    infrastructure planning.

    Parameters
    ----------
    return_periods : list of str
        Return periods to extract. Options: rp10, rp25, rp50,
        rp100, rp250, rp500, rp1000
    hazard_type : str
        Type of flood hazard. Options: inunriver, inuncoast
    buffer_meters : int
        Buffer radius around each school point in metres
    climate_scenario : str
        Climate scenario. Options: historical, rcp4p5, rcp8p5
    data_dir : str or Path
        Directory to store downloaded raster files

    Examples
    --------
    >>> handler = AQUEDUCTFloodHandler(
    ...     return_periods=["rp100", "rp1000"],
    ...     hazard_type="inunriver",
    ...     buffer_meters=500
    ... )
    >>> schools_gdf = gpd.GeoDataFrame(...)
    >>> risk_df = handler.get_flood_risk(schools_gdf)
    """

    return_periods: List[str] = None
    hazard_type: Literal["inunriver", "inuncoast"] = "inunriver"
    buffer_meters: int = 500
    climate_scenario: Literal[
        "historical", "rcp4p5", "rcp8p5"
    ] = "historical"
    data_dir: Optional[Union[str, Path]] = None

    def __post_init__(self):
        """Initialise handler with defaults and validation."""
        if self.return_periods is None:
            self.return_periods = ["rp100"]

        # Validate return periods
        for rp in self.return_periods:
            if rp not in RETURN_PERIODS:
                raise ValueError(
                    f"Invalid return period: {rp}. "
                    f"Valid options: {list(RETURN_PERIODS.keys())}"
                )

        # Validate hazard type
        if self.hazard_type not in HAZARD_TYPES:
            raise ValueError(
                f"Invalid hazard type: {self.hazard_type}. "
                f"Valid options: {list(HAZARD_TYPES.keys())}"
            )

        # Set data directory
        if self.data_dir is None:
            self.data_dir = Path(tempfile.mkdtemp())
        else:
            self.data_dir = Path(self.data_dir)
            self.data_dir.mkdir(parents=True, exist_ok=True)

        logger.info(
            f"AQUEDUCTFloodHandler initialised: "
            f"{self.hazard_type} | "
            f"Return periods: {self.return_periods} | "
            f"Scenario: {self.climate_scenario}"
        )

    def _get_filename(self, return_period: str) -> str:
        """Generate AQUEDUCT filename for given return period."""
        scenario_code = {
            "historical": "0",
            "rcp4p5":     "4",
            "rcp8p5":     "8"
        }[self.climate_scenario]

        return (
            f"{self.hazard_type}_"
            f"{return_period}_"
            f"{scenario_code}_"
            f"perc_50.tif"
        )

    def _get_download_url(self, return_period: str) -> str:
        """Get AQUEDUCT download URL for given return period."""
        filename = self._get_filename(return_period)
        base_url = (
            "https://wri-public-data.s3.amazonaws.com/"
            "resourcewatch/foo_024_aqueduct_global_flood_risk/"
        )
        return base_url + filename

    def _download_raster(self, return_period: str) -> Path:
        """
        Download AQUEDUCT raster for given return period.

        Parameters
        ----------
        return_period : str
            Return period identifier (e.g. rp100)

        Returns
        -------
        Path
            Path to downloaded raster file
        """
        filename = self._get_filename(return_period)
        filepath = self.data_dir / filename

        if filepath.exists():
            logger.info(f"Using cached raster: {filename}")
            return filepath

        url = self._get_download_url(return_period)
        logger.info(f"Downloading {filename} from WRI...")

        try:
            r = requests.get(url, stream=True, timeout=60)
            r.raise_for_status()
            with open(filepath, "wb") as f:
                for chunk in r.iter_content(chunk_size=8192):
                    f.write(chunk)
            logger.info(f"Downloaded: {filename}")
            return filepath

        except requests.exceptions.RequestException as e:
            raise ConnectionError(
                f"Failed to download AQUEDUCT data: {e}. "
                f"Please download manually from: {url}"
            )

    def _extract_values(
        self,
        raster_path: Path,
        schools_gdf: gpd.GeoDataFrame,
        return_period: str
    ) -> pd.Series:
        """
        Extract raster values at school locations with buffer.

        Parameters
        ----------
        raster_path : Path
            Path to AQUEDUCT raster file
        schools_gdf : GeoDataFrame
            School locations in EPSG:4326
        return_period : str
            Return period identifier for column naming

        Returns
        -------
        Series
            Flood depth values per school
        """
        values = []

        with rasterio.open(raster_path) as src:
            # Reproject schools to match raster CRS if needed
            if schools_gdf.crs != src.crs:
                schools_proj = schools_gdf.to_crs(src.crs)
            else:
                schools_proj = schools_gdf.copy()

            # Buffer schools
            buffer_deg = self.buffer_meters / 111000
            schools_proj["geometry"] = schools_proj.geometry.buffer(
                buffer_deg
            )

            for _, school in schools_proj.iterrows():
                try:
                    geom = [mapping(school.geometry)]
                    out_image, _ = mask(src, geom, crop=True)
                    data = out_image[0]

                    # Replace nodata with nan
                    nodata = src.nodata
                    if nodata is not None:
                        data = data.astype(float)
                        data[data == nodata] = np.nan

                    # Mean flood depth in buffer
                    mean_val = np.nanmean(data)
                    values.append(
                        0.0 if np.isnan(mean_val) else round(mean_val, 3)
                    )

                except Exception:
                    values.append(0.0)

        col_name = f"flood_depth_{return_period}_m"
        return pd.Series(values, name=col_name)

    def get_flood_risk(
        self,
        schools: Union[gpd.GeoDataFrame, pd.DataFrame],
        lat_col: str = "latitude",
        lon_col: str = "longitude",
        id_col: Optional[str] = None
    ) -> pd.DataFrame:
        """
        Extract flood hazard values for school locations.

        Parameters
        ----------
        schools : GeoDataFrame or DataFrame
            School locations. If DataFrame, must have lat/lon columns.
        lat_col : str
            Name of latitude column (if DataFrame input)
        lon_col : str
            Name of longitude column (if DataFrame input)
        id_col : str, optional
            Name of school ID column to include in output

        Returns
        -------
        DataFrame
            School-level flood depth for each return period,
            plus a composite risk score and risk tier.

        Examples
        --------
        >>> import pandas as pd
        >>> schools = pd.DataFrame({
        ...     "school_id": ["S001", "S002"],
        ...     "latitude":  [10.76,  11.93],
        ...     "longitude": [79.84,  79.83]
        ... })
        >>> handler = AQUEDUCTFloodHandler(return_periods=["rp100"])
        >>> result = handler.get_flood_risk(schools, id_col="school_id")
        """
        # Convert DataFrame to GeoDataFrame if needed
        if isinstance(schools, pd.DataFrame):
            geometry = [
                Point(lon, lat)
                for lat, lon in zip(
                    schools[lat_col], schools[lon_col]
                )
            ]
            schools_gdf = gpd.GeoDataFrame(
                schools, geometry=geometry, crs="EPSG:4326"
            )
        else:
            schools_gdf = schools.copy()
            if schools_gdf.crs is None:
                schools_gdf = schools_gdf.set_crs("EPSG:4326")

        logger.info(
            f"Extracting flood risk for "
            f"{len(schools_gdf)} schools..."
        )

        # Build result dataframe
        result = pd.DataFrame()

        # Add ID column if specified
        if id_col and id_col in schools_gdf.columns:
            result[id_col] = schools_gdf[id_col].values

        # Add coordinates
        if hasattr(schools_gdf.geometry, "x"):
            result["longitude"] = schools_gdf.geometry.x.values
            result["latitude"]  = schools_gdf.geometry.y.values

        # Extract values for each return period
        for rp in self.return_periods:
            logger.info(f"Processing return period: {rp}")
            try:
                raster_path = self._download_raster(rp)
                depths = self._extract_values(
                    raster_path, schools_gdf, rp
                )
                result[f"flood_depth_{rp}_m"] = depths.values
            except Exception as e:
                logger.warning(
                    f"Failed to process {rp}: {e}. "
                    f"Setting to 0."
                )
                result[f"flood_depth_{rp}_m"] = 0.0

        # Composite risk score
        depth_cols = [
            c for c in result.columns
            if c.startswith("flood_depth_")
        ]
        if depth_cols:
            result["max_flood_depth_m"] = result[depth_cols].max(axis=1)
            result["mean_flood_depth_m"] = result[depth_cols].mean(
                axis=1
            ).round(3)
            result["flood_risk_score"] = (
                result["max_flood_depth_m"] /
                max(result["max_flood_depth_m"].max(), 0.01) * 100
            ).round(1)
            result["flood_risk_tier"] = result[
                "flood_risk_score"
            ].apply(self._assign_risk_tier)

        result["hazard_type"]       = self.hazard_type
        result["climate_scenario"]  = self.climate_scenario
        result["return_periods"]    = str(self.return_periods)

        logger.info(
            f"Flood risk extraction complete. "
            f"CRITICAL: "
            f"{(result.flood_risk_tier == 'CRITICAL').sum()}, "
            f"HIGH: "
            f"{(result.flood_risk_tier == 'HIGH').sum()}"
        )

        return result

    @staticmethod
    def _assign_risk_tier(score: float) -> str:
        """
        Assign risk tier based on normalised flood risk score.

        Parameters
        ----------
        score : float
            Normalised flood risk score (0-100)

        Returns
        -------
        str
            Risk tier: CRITICAL, HIGH, MEDIUM, or LOW
        """
        if score >= 75:   return "CRITICAL"
        elif score >= 50: return "HIGH"
        elif score >= 25: return "MEDIUM"
        else:             return "LOW"

    def get_dataset_info(self) -> dict:
        """
        Return metadata about the AQUEDUCT dataset.

        Returns
        -------
        dict
            Dataset metadata including source, coverage and resolution
        """
        return {
            "name":         "WRI AQUEDUCT 3.0 Flood Hazard",
            "source":       "World Resources Institute",
            "url":          "https://www.wri.org/aqueduct",
            "hazard_type":  self.hazard_type,
            "description":  HAZARD_TYPES[self.hazard_type],
            "scenario":     self.climate_scenario,
            "return_periods": {
                rp: RETURN_PERIODS[rp]
                for rp in self.return_periods
            },
            "resolution":   "~1km (30 arc-seconds)",
            "coverage":     "Global",
            "units":        "Flood inundation depth (metres)",
            "license":      "CC BY 4.0",
            "citation": (
                "Ward et al. (2020). Aqueduct Floods Methodology. "
                "World Resources Institute, Washington, DC."
            )
        }

    def __repr__(self) -> str:
        return (
            f"AQUEDUCTFloodHandler("
            f"hazard_type={self.hazard_type!r}, "
            f"return_periods={self.return_periods!r}, "
            f"scenario={self.climate_scenario!r}, "
            f"buffer_meters={self.buffer_meters})"
        )
'''

# Write to file
os.makedirs('gigaspatial/handlers', exist_ok=True)
with open('gigaspatial/handlers/aqueduct_flood.py', 'w') as f:
    f.write(flood_handler_code)

print('✅ AQUEDUCTFloodHandler written!')
print('File: gigaspatial/handlers/aqueduct_flood.py')
print(f'Lines: {len(flood_handler_code.splitlines())}')
```
