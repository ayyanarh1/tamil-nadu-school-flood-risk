```python
# Phase 2 PDF report
from fpdf import FPDF

class Report(FPDF):
    def header(self):
        pass

    def footer(self):
        self.set_y(-15)
        self.set_font('Helvetica', 'I', 8)
        self.set_text_color(128)
        self.cell(0, 10,
            'Page ' + str(self.page_no()) +
            ' | github.com/ayyanarh1/tamil-nadu-school-flood-risk',
            align='C')

    def section_title(self, title):
        self.set_font('Helvetica', 'B', 13)
        self.set_text_color(6, 90, 130)
        self.set_fill_color(232, 244, 253)
        self.cell(0, 9, title, fill=True)
        self.ln(5)
        self.set_text_color(0)

    def body_text(self, text):
        self.set_font('Helvetica', '', 11)
        self.set_text_color(50)
        self.multi_cell(0, 6, text)
        self.ln(3)

    def bullet(self, text):
        self.set_font('Helvetica', '', 11)
        self.set_text_color(50)
        self.set_x(15)
        self.multi_cell(0, 6, '- ' + text)

pdf = Report()
pdf.set_auto_page_break(auto=True, margin=15)
pdf.add_page()

# Title
pdf.set_font('Helvetica', 'B', 22)
pdf.set_text_color(2, 28, 59)
pdf.cell(0, 12, 'Multi-Country School Flood Risk Analysis', align='C')
pdf.ln(8)
pdf.set_font('Helvetica', 'B', 14)
pdf.set_text_color(6, 90, 130)
pdf.cell(0, 8, 'Tamil Nadu, India + Mozambique', align='C')
pdf.ln(6)
pdf.set_font('Helvetica', '', 11)
pdf.set_text_color(100)
pdf.cell(0, 6, 'UNICEF Digital Inclusion - Giga Initiative | March 2026', align='C')
pdf.ln(10)

# Section 1
pdf.section_title('1. Executive Summary')
pdf.body_text(
    'This report presents a multi-country satellite-based flood risk assessment '
    'covering 25 schools across Tamil Nadu (India) and Mozambique, combining '
    '6 data sources spanning 40 years of climate and flood observations.'
)
pdf.body_text(
    'Key finding: 16 of 25 schools (64%) are at CRITICAL risk. Both countries '
    'show the same pattern - coastal schools with no connectivity face compound '
    'risk from flood and cyclone hazards. The pipeline is generalisable across '
    'LMIC contexts.'
)

# Section 2
pdf.section_title('2. Countries Covered')
pdf.body_text('Tamil Nadu, India:')
pdf.bullet('15 schools analysed across 12 districts')
pdf.bullet('10 CRITICAL, 4 HIGH, 1 MEDIUM risk schools')
pdf.bullet('7 schools with no internet connectivity')
pdf.bullet('Peak flood season: Northeast monsoon Oct-Dec')
pdf.ln(4)
pdf.body_text('Mozambique:')
pdf.bullet('10 schools analysed across 9 provinces')
pdf.bullet('6 CRITICAL, 2 HIGH, 1 MEDIUM, 1 LOW risk schools')
pdf.bullet('7 schools with no internet connectivity')
pdf.bullet('Peak hazard: Cyclone season Nov-Apr')
pdf.ln(4)

# Section 3
pdf.section_title('3. Risk Framework')
pdf.body_text('Full IPCC H x E x V framework applied:')
pdf.bullet('Hazard (40%) - SAR + JRC + GFD + ERA5 + IBTrACS cyclone tracks')
pdf.bullet('Exposure (30%) - Coastal location + flood zone classification')
pdf.bullet('Vulnerability (30%) - Connectivity + hospital distance + rural')
pdf.ln(4)

# Section 4
pdf.add_page()
pdf.section_title('4. Tamil Nadu Key Findings')
pdf.body_text('CRITICAL schools requiring immediate action:')
critical_tn = [
    'School Puducherry Border (68.7) - Near permanent water presence',
    'Panchayat School Nagapattinam (68.3) - No connectivity + high rainfall',
    'School Ramanathapuram (67.0) - No connectivity + 94km from hospital',
    'School Kanchipuram (66.6) - No connectivity + high cyclone exposure',
    'Panchayat School Tirunelveli (65.9) - 136km from hospital',
    'Govt School Cuddalore (65.6) - No connectivity + high cyclone',
    'School Villupuram (62.3) - No connectivity + high cyclone',
    'School Tuticorin (54.6) - Coastal + flood exposed',
    'Govt High School Chennai (44.8) - Highest cyclone exposure',
    'Govt School Tiruchirappalli (41.1) - No connectivity',
]
for s in critical_tn:
    pdf.bullet(s)
pdf.ln(4)

# Section 5
pdf.section_title('5. Mozambique Key Findings')
pdf.body_text('CRITICAL schools requiring immediate action:')
critical_moz = [
    'School Inhambane Coast (100.2) - Extreme flood + cyclone exposure',
    'Primary School Quelimane (86.2) - No connectivity + high cyclone',
    'Secondary School Beira Coast (78.3) - Cyclone Idai epicentre',
    'Primary School Sofala (73.3) - No connectivity + high cyclone',
    'Primary School Tete (61.3) - No connectivity + flood exposed',
    'Primary School Gaza (53.8) - Flood + cyclone compound risk',
]
for s in critical_moz:
    pdf.bullet(s)
pdf.ln(4)

# Section 6
pdf.section_title('6. Comparative Analysis')
pdf.body_text('Key similarities between Tamil Nadu and Mozambique:')
pdf.bullet('Both show coastal school concentration of risk')
pdf.bullet('Both have 7 schools with no internet connectivity')
pdf.bullet('Both show cyclone tracks compounding flood risk')
pdf.bullet('Inland schools significantly safer in both countries')
pdf.ln(4)
pdf.body_text('Key differences:')
pdf.bullet('Mozambique has higher cyclone intensity (Indian Ocean vs Bay of Bengal)')
pdf.bullet('Tamil Nadu has higher rainfall-driven flood risk')
pdf.bullet('Mozambique schools more isolated (hospital distance)')
pdf.ln(4)

# Section 7
pdf.section_title('7. 2050 Climate Projections')
pdf.body_text('Under SSP5-8.5 worst case scenario by 2050:')
pdf.bullet('Tamil Nadu: 5 schools escalate to higher risk tier')
pdf.bullet('3 new CRITICAL schools: Ramanathapuram, Cuddalore, Tuticorin')
pdf.bullet('Mozambique: similar escalation expected under SSP5-8.5')
pdf.bullet('Climate action (SSP2-4.5) reduces escalating schools by 40%')
pdf.ln(4)

# Section 8
pdf.section_title('8. Recommendations')
recs = [
    'URGENT: Connectivity investment for Nagapattinam and Quelimane - highest compound risk',
    'URGENT: Emergency access planning for Tirunelveli and Inhambane - most isolated',
    'MONITOR: Real-time Sentinel-1 alerts during peak hazard seasons',
    'PLAN: Integrate 2050 projections into school infrastructure planning',
    'EXPAND: Apply pipeline to full Giga dataset across both countries',
    'REPLICATE: Pipeline tested in 2 LMIC contexts - ready for global scaling',
]
for r in recs:
    pdf.bullet(r)
pdf.ln(6)

pdf.set_font('Helvetica', 'I', 10)
pdf.set_text_color(100)
pdf.cell(0, 6,
    'Full code: github.com/ayyanarh1/tamil-nadu-school-flood-risk',
    align='C')

pdf.output('multi_country_risk_report.pdf')
print('✅ Phase 2 PDF report saved!')
print('Pages:', pdf.page_no())
```
